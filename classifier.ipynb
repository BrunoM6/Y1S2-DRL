# Data Loading and Analysis

```python
import os
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from dataset import ImageBatchDataset
from sklearn.model_selection import train_test_split
from torch.utils.data import Subset

FOLDERS = ["inpainting", "insight", "text2img", "wiki"]
DATA_DIR = "data"
BATCH_SIZE = 32

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((512, 512))
])

dataset = ImageBatchDataset(FOLDERS, DATA_DIR, transform=transform)
print(f"Found images: {len(dataset)}")

targets = [label for _, label, _ in dataset.file_paths]

train_idx, test_idx = train_test_split(
    range(len(dataset)),
    test_size=0.2,
    random_state=42,
    stratify=targets
)

train_dataset = Subset(dataset, train_idx)
test_dataset = Subset(dataset, test_idx)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"Training on: {len(train_dataset)} images")
print(f"Testing on: {len(test_dataset)} images")
```
