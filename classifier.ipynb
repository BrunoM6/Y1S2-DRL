# Data Loading and Analysis

```python
import os
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from cnn import ImageBatchDataset

FOLDERS = ["inpainting", "insight", "text2img", "wiki"]
DATA_DIR = "data"
BATCH_SIZE = 32

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((512, 512))
])

dataset = ImageBatchDataset(FOLDERS, DATA_DIR, transform=transform)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=0)

print(f"Found images: {len(dataset)}")

for i, (images, folder) in enumerate(loader):
    print(f"Batch {i}: with {len(images)} images.")
```
